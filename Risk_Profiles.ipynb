```python
import matplotlib.pyplot as plt

#1 Define the points
points = [(3.8, 0.25), (3.6, 0.5), (3.5, 0.25) ]

# Separate x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, color='blue')
plt.xlabel('Profit (million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile - Inventory & Supplier Diversification')

# Draw lines from x-axis to points
for point in points:
    plt.hlines(point[1], 0, point[0], linestyle='dotted', colors='gray')
    plt.vlines(point[0], 0, point[1], linestyle='solid', colors='gray')

# Set x-axis limit
plt.xlim(3, 4)

# Show grid
plt.grid(True)

# Show plot
plt.show()

#2 Define the points
points = [(3.88, 0.20), (3.68, 0.5), (3.40, 0.30) ]

# Separate x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, color='blue')
plt.xlabel('Profit (million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile - Labor Training')

# Draw lines from x-axis to points
for point in points:
    plt.hlines(point[1], 0, point[0], linestyle='dotted', colors='gray')
    plt.vlines(point[0], 0, point[1], linestyle='solid', colors='gray')

# Set x-axis limit
plt.xlim(3, 4)

# Show grid
plt.grid(True)

# Show plot
plt.show()

#3 Define the points
points = [(3.95, 0.25), (3.40, 0.5), (3.20, 0.25) ]

# Separate x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, color='blue')
plt.xlabel('Profit (million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile - Contingency Financing')

# Draw lines from x-axis to points
for point in points:
    plt.hlines(point[1], 0, point[0], linestyle='dotted', colors='gray')
    plt.vlines(point[0], 0, point[1], linestyle='solid', colors='gray')

# Set x-axis limit
plt.xlim(3, 4)

# Show grid
plt.grid(True)

# Show plot
plt.show()

#4 Define the points
points = [(3.82, 0.25), (3.80, 0.5), (3.68, 0.25) ]

# Separate x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, color='blue')
plt.xlabel('Profit (million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile - Up-Front Planning Improvement')

# Draw lines from x-axis to points
for point in points:
    plt.hlines(point[1], 0, point[0], linestyle='dotted', colors='gray')
    plt.vlines(point[0], 0, point[1], linestyle='solid', colors='gray')

# Set x-axis limit
plt.xlim(3, 4)

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
